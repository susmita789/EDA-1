```python
df['camly'] = df.mention.cumsum()
```

```python
def logarithm(x):
    b = np.log(x)
    return b
df['logarithm'] = df['camly'].apply(logarithm)
df['square'] = df['mention'].apply(lambda x : x**2)
```
