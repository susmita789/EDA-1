# visualisation using seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x = 'size' , y = 'total_bill',data = tips,hue = 'smoker')
```

```python
sns.relplot(x = 'size' , y = 'total_bill',data = tips,hue = 'time')
```

```python
tips
```

```python
sns.relplot(x = 'size' , y = 'total_bill',data = tips,hue = 'day',style = 'time')
```

```python
sns.relplot(x = 'tip',y = 'total_bill',data = tips)
```

```python
# separating the data
sns.relplot(x = 'tip',y = 'total_bill',data = tips,hue = 'smoker')
```

```python
# separating the data
sns.relplot(x = 'tip',y = 'total_bill',data = tips,hue = 'day')
```

```python
#every thursday member comming is smoker or non smoker -------------grouping
sns.relplot(x = 'tip',y = 'total_bill',data = tips,hue = 'day',style = 'smoker')
```

```python
sns.relplot(x = 'tip',y = 'total_bill',data = tips,hue = 'day',style = 'smoker',size = 'size')
```

```python
sns.relplot(x = 'tip',y = 'total_bill',data = tips ,size = 'size')
```
